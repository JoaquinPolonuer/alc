```python
import numpy as np
import matplotlib.pyplot as plt
```

### Ejercicio 12

```python
n = 3
A = np.random.rand(n,n)
suc = [0]

for i in range(100):
    x = np.random.rand(n) - 0.5
    x = x/np.linalg.norm(x, 2)*np.random.rand()
    Ax = A@x
    norm = np.linalg.norm(Ax, 2) / np.linalg.norm(x,2)
    
    suc.append(max(suc[-1], norm))
    
norm = np.linalg.norm(A, 2)
plt.plot(suc)
plt.plot([norm]*len(suc))

print("La norma calculada por la sucesion es", suc[-1])
print("La norma real es", norm)
```

### Ejercicio 14

```python
def generate_random_error(const):
    e = np.random.rand(3)
    e = e / np.linalg.norm(e, np.inf)
    e = e *const 
    #print(np.linalg.norm(e, np.inf))
    return e
```

```python
A = np.array([[3,0,0],[0,5/4, 3/4],[0, 3/4, 5/4]])
cond_a = np.linalg.cond(A, np.inf)

b = np.array([3,2,2])
norm_b = np.linalg.norm(b, np.inf)
```

```python
errors = []
for i in range(1000):
    norm_e = 10e-4*norm_b/cond_a
    b_ = b+generate_random_error(norm_e)
    
    x = np.linalg.solve(A,b)
    x_ = np.linalg.solve(A,b_)

    relative_error = np.linalg.norm(x - x_, np.inf) / np.linalg.norm(x, np.inf)
    errors.append(relative_error)

plt.plot(errors)
plt.plot([10e-4]*len(errors))
```
